```python
pd.crosstab(df['y'], df['x.radius_mean']>15, normalize='index')
```

```python
# P(y = M)
p_y_M = (df['y'] == 'M').mean()

# P(radius > 15 | y = M)
p_r_given_M = df[df['y']=='M']['x.radius_mean'].gt(15).mean()

# P(radius > 15)
p_r = df['x.radius_mean'].gt(15).mean()

# Bayes
p_M_given_r = (p_r_given_M * p_y_M) / p_r
```
